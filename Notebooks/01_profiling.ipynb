```python
# ============================================================
# NOTEBOOK 01 — DATA PROFILING
# Think of this like a health checkup for your data.
# A doctor checks your heart, blood, eyes.
# We check nulls, duplicates, formats, weird values.
# ============================================================

# Cell 1 — Import tools we need
# pandas = our main tool for reading + working with data
# os = helps us find files on the computer

import pandas as pd
import os

print("✅ Libraries loaded!")
```

```python
# Loading the CSV from your Data folder on Desktop

df = pd.read_csv("../Data/raw_hr_data.csv")

print(f"📊 Dataset has {df.shape[0]} rows and {df.shape[1]} columns")
```

```python
df.head
```

```python
# Cell 4 — See all column names
# These are the "headers" of each column, like column A, B, C in Excel
# We need to know what fields exist before we can fix them

print("📋 Column names in this dataset:")
print(df.columns.tolist())
```

```python
# Cell 5 — Check data types
# Every column holds a TYPE of data:
#   int64   = whole numbers (like age: 34)
#   object  = text / strings (like department: "Sales")
#   float64 = decimal numbers (like salary: 50000.50)
# Wrong types = problems later. We must catch them now.

print("🔍 Data types per column:")
print(df.dtypes)
```

```python
# Cell 6 — Find NULL / missing values
# A null = an empty cell. Like a form with a blank field.
# .isnull().sum() counts how many blanks exist in each column
# We only show columns that actually HAVE nulls

null_counts = df.isnull().sum()
null_columns = null_counts[null_counts > 0]

if null_columns.empty:
    print("✅ No missing values found!")
else:
    print("⚠️ Columns with missing values:")
    print(null_columns)
```

```python
# Cell 7 — Find duplicate rows
# A duplicate = the same employee appearing twice
# That's a data entry mistake. We must count them.
# .duplicated() marks rows that are exact copies
# .sum() counts how many there are

duplicate_count = df.duplicated().sum()
print(f"🔁 Number of duplicate rows: {duplicate_count}")

# If there are duplicates, show them
if duplicate_count > 0:
    print("\nHere are the duplicate rows:")
    print(df[df.duplicated(keep=False)])
```

```python
# Cell 8 — Check the range of numerical columns
# .describe() gives us stats like min, max, average
# We use this to spot WEIRD values
# Example: if Age has a min of -5, that's a mistake!

print("📈 Summary statistics for number columns:")
df.describe()
```

```python
# Cell 9 — Check unique values in category columns
# Categorical columns = columns with a small set of options
# Like: Gender can only be Male or Female
# If we see "male", "MALE", "M" — that's a mess we need to fix

# These are the columns we expect to have fixed options
category_columns = [
    "Attrition", "BusinessTravel", "Department",
    "EducationField", "Gender", "JobRole",
    "MaritalStatus", "OverTime"
]

print("🗂️ Unique values in category columns:\n")
for col in category_columns:
    if col in df.columns:
        unique_vals = df[col].unique()
        print(f"  {col}: {unique_vals}")
        print()
```

```python
# Cell 10 — Check for constant columns
# A constant column = same value in every single row
# Example: "StandardHours" is 80 for everyone — useless!
# These columns add no information. We'll flag them.

print("🚩 Columns with only ONE unique value (probably useless):")
for col in df.columns:
    if df[col].nunique() == 1:
        print(f"  ⛔ {col} — value is always: {df[col].unique()[0]}")
```

```python
# Cell 11 — Build a PROFILING REPORT
# This is our summary doc — like a doctor's report after a checkup
# We collect everything we found and save it

report = []

# --- Section 1: Basic info ---
report.append("=" * 50)
report.append("HR DATA PROFILING REPORT")
report.append("=" * 50)
report.append(f"Total rows    : {df.shape[0]}")
report.append(f"Total columns : {df.shape[1]}")
report.append("")

# --- Section 2: Missing values ---
report.append("--- MISSING VALUES ---")
null_counts = df.isnull().sum()
null_found = null_counts[null_counts > 0]
if null_found.empty:
    report.append("No missing values.")
else:
    for col, count in null_found.items():
        report.append(f"  {col}: {count} missing")
report.append("")

# --- Section 3: Duplicates ---
report.append("--- DUPLICATES ---")
report.append(f"Duplicate rows: {df.duplicated().sum()}")
report.append("")

# --- Section 4: Constant columns ---
report.append("--- CONSTANT COLUMNS (to drop) ---")
for col in df.columns:
    if df[col].nunique() == 1:
        report.append(f"  {col} = always '{df[col].unique()[0]}'")
report.append("")

# --- Section 5: Data types ---
report.append("--- DATA TYPES ---")
for col, dtype in df.dtypes.items():
    report.append(f"  {col}: {dtype}")

# Print the report
print("\n".join(report))
```

```python
# Cell 12 — Save the report to a text file
# So we have a record of all problems found

report_path = os.path.expanduser("../Data/raw_hr_data.csv")

with open(report_path, "w") as f:
    f.write("\n".join(report))

print(f"💾 Report saved to: {report_path}")
print("✅ Notebook 01 complete. Move to 02_mapping_cleaning.ipynb next.")
```
